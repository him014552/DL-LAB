```python
# Q.1. Plot the following activation functions and their derivatives in Python.
```

```python
#Linear Activation Function

x = np.linspace(-10, 10, 400)
def linear(x):
    return x
def linear_derivative(x):
    return np.ones_like(x)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, linear(x), label="Linear")
plt.title("Linear Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, linear_derivative(x), label="Derivative", linestyle='--')
plt.title("Linear Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#Sigmoid Activation Function
 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.title("Sigmoid Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_derivative(x), label="Derivative", linestyle='--')
plt.title("Sigmoid Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#Tanh Activation Function 

def tanh(x):
    return np.tanh(x)
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, tanh(x), label="Tanh")
plt.title("Tanh Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, tanh_derivative(x), label="Derivative", linestyle='--')
plt.title("Tanh Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#Softmax Activation Function (for simplicity, assuming x is a single vector) 

def softmax(x):
    exps = np.exp(x - np.max(x))  # For numerical stability
    return exps / np.sum(exps)
def softmax_derivative(x):+
    
    s = softmax(x)
    return s * (1 - s)  # Simplified derivative (diag(s) - s * s.T) for 1D case

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, softmax(x), label="Softmax")
plt.title("Softmax Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, softmax_derivative(x), label="Derivative", linestyle='--')
plt.title("Softmax Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#ReLU Activation Function

def relu(x):
    return np.maximum(0, x)
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, relu(x), label="ReLU")
plt.title("ReLU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, relu_derivative(x), label="Derivative", linestyle='--')
plt.title("ReLU Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#Leaky ReLU Activation Function

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.title("Leaky ReLU Activation Function")
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, leaky_relu_derivative(x), label="Derivative", linestyle='--')
plt.title("Leaky ReLU Derivative")
plt.legend()
plt.grid()
plt.show()
```

#Q.2. Take training data and make dataframe and the data should have null values. Make dummy data using pandas, then normalize the data  and remove null values and do data inquisition and use min-max normalization. 
   Make sequential neural network (2 hidden layers) using Keras and Tensorflow.

```python
import pandas as pd
import numpy as np

# Create a DataFrame with dummy data
data = {
    'f1': [10, 20, 30, np.nan, 50],
    'f2': [5, 15, np.nan, 35, 45],
    'f3': [100, np.nan, 300, 400, 500],
    'f4': [1, 2, 3, 4, np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Handle null values (replace with column mean)
df_filled = df.fillna(df.mean())

# Min-Max Normalization
normalized_df = (df_filled - df_filled.min()) / (df_filled.max() - df_filled.min())
print("\nNormalized DataFrame:")
print(normalized_df)
```

```python
# Check for null values
print("\nNull Values in the DataFrame:")
print(df.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Prepare data for training
X = normalized_df.values  # Features (input data)
y = np.array([0, 1, 1, 0, 1])  # Labels (target values)

# Create a simple sequential neural network
model = Sequential()

# Add layers to the model
model.add(Dense(16, activation='relu', input_dim=X.shape[1]))  # First hidden layer with 16 neurons
model.add(Dense(8, activation='relu'))  # Second hidden layer with 8 neurons
model.add(Dense(1, activation='sigmoid'))  # Output layer (for binary classification)

# Compile the model (set up optimization and evaluation metrics)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=10, batch_size=2, verbose=1)

# Show a summary of the model
model.summary()
```
